### Audio amplifier characterization

```python
import numpy as np 
import plotly.graph_objs as go
 
v_a = np.array([5.4, 4.5, 3.6, 2.7, 1.8, 0.9, 0, -0.9])
vpp_signal = np.array([9.04, 7.29, 5.47, 3.81, 1.92, 0.994, 0.498, 0.53]) 
vpp_audio = np.array([1.51, 2.9, 5.37, 9.03, 9.86, 4.7, 8.43, 10.94]) 
vrms_volumne = np.array([0.149, 0.527, 1.31, 2.48, 2.79, 2.89, 3.49, 2.21]) 

v_c = v_a*(33/593)
gain_audio = 20*np.log10(vpp_audio/vpp_signal)

# Create figure with secondary y-axis
fig = go.Figure()

fig.add_trace(go.Scatter(x=v_c, y=gain_audio, mode='markers', name='Gain (dB)')) 
fig.update_layout(xaxis_title='V<sub>C</sub>', yaxis_title='Gain (dB) at V<sub>audio</sub>') 

coef = np.polyfit(v_c, gain_audio, deg=1) # deg=1 for order 1 polynomial (linear)
fit = coef[0]*v_c + coef[1]
fig.add_trace(go.Scatter(x=v_c, y=fit, mode='lines', line=dict(dash='dash'), name='Best fit for Gain at V<sub>audio</sub>'))
```

**Plot the graph of the gain versus 𝑉c, and discuss whether the output performance of the audio amplifier system meets the desired objective**

Based on the results, the lower Vc would produce a higher gain. Which would mean that the audio produce is louder. This aligns with Lab 5.
